```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
target = 'vacc_h1n1_f'
# target = 'vacc_seas_f'
train = pd.merge(pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train.csv'), 
                 pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train_labels.csv')[target], left_index=True, right_index=True)
test = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/test.csv')
sample_submission = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/submission.csv')
```

```python
#train = train.dropna(thresh = 20, axis = 0)
```

```python
X = train.drop(columns = target)
y = train[target]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, train_size=0.8, random_state= 30)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from category_encoders import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline


pipe = make_pipeline(
    OneHotEncoder(use_cat_names=True),  
    SimpleImputer(), 
    DecisionTreeClassifier(random_state=42, criterion='entropy')
)

pipe.fit(X_train, y_train)
print('훈련 정확도: ', pipe.score(X_train, y_train))
print('검증 정확도: ', pipe.score(X_val, y_val))
```

```python
import numpy as np

def engineer(df):
    """특성을 엔지니어링 하는 함수입니다."""
    
    # 높은 카디널리티를 가지는 특성을 제거합니다.
    selected_cols = df.select_dtypes(include=['number', 'object'])
    labels = selected_cols.nunique() # 특성별 카디널리티 리스트
    selected_features = labels[labels <= 30].index.tolist() # 카디널리티가 30보다 작은 특성만 선택합니다.
    df = df[selected_features]
    
    # 새로운 특성을 생성합니다.
    behaviorals = [col for col in df.columns if 'behavioral' in col] 
    df['behaviorals'] = df[behaviorals].sum(axis=1)
    
    
    dels = [col for col in df.columns if ('employment' in col or 'seas' in col)]
    df.drop(columns=dels, inplace=True)
        
    return df
```

**Simple Imputer**  
:결측값을 대입하기 위한 기본 전략을 제공한다. 결측값은 제공된 상수 값으로 대치되거나 결측값이 있는 각 열의 통계(mean, median, most_frequent)를 사용하여 대치될 수 있다.

```python
from sklearn.impute import SimpleImputer
from sklearn.metrics import f1_score
import numpy as np

pipe = make_pipeline(
    OneHotEncoder(use_cat_names=True), 
    SimpleImputer(missing_values = np.nan, strategy = 'mean'), # missing_values 열의 평균값을 대체하는 방법을 보여준다. default값
    DecisionTreeClassifier(min_samples_leaf=8, max_depth = 8, random_state=30, criterion='gini')
)

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_val)
f1 = f1_score(y_val, y_pred) # Val f1_score
print('훈련 정확도', pipe.score(X_train, y_train))
print('검증 정확도', pipe.score(X_val, y_val))
print('Val f1 score :',f1)
```

**Iterative Imputer**   
: 다른 모든 것에서 각 특성을 추정하는 다변수 대입기. 라운드 로빈 방식으로 다른 기능의 함수로 결측값이 있는 각 기능을 모델링하여 결측값을 대치하는 전략입니다. 각 단계에서 기능 열은 출력으로 지정되고 y다른 기능 열은 입력으로 처리됩니다 X. 회귀자는 알려진 에 적합합니다 . 그런 다음 회귀자를 사용하여 의 결측값을 예측합니다 .

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_iterative_imputer
# now you can import normally from sklearn.impute
from sklearn.impute import IterativeImputer
from sklearn.metrics import f1_score

pipe = make_pipeline(
    OneHotEncoder(use_cat_names=True), 
    IterativeImputer(max_iter=10, random_state=30), 
    # 이것은 반복적인 방식으로 각 기능에 대해 수행된 다음 대치 라운드에 대해 반복됩니다. 최종 전가 라운드의 결과가 반환됩니다.(X, y)yymax_iter
    DecisionTreeClassifier(min_samples_leaf=8, max_depth = 8, random_state=30, criterion='gini')
)

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_val)
f1 = f1_score(y_val, y_pred)
print('훈련 정확도', pipe.score(X_train, y_train))
print('검증 정확도', pipe.score(X_val, y_val))
print('Val f1 score :',f1)# Val f1_score
```
